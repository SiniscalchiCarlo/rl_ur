# SARSA for the Windy Gridworld problem

In this exercise you will have to implement the SARSA algorithm on the Windy Gridworld problem. The Windy Gridworld problem is given in Example 6.5 on page 130 in the book. We will implement the example and exercise 6.9 which is on the next page. In the Windy Gridworld, the agent has to reach the goal state from the starting state, while taking one of the four actions: up, right down and left. In exercise 6.9 these actions are extended to include the diagonal ones, leading to eight actions in total. In the Windy Gridworld some states experience wind that push the agent upwards.

```python
import gymnasium as gym
import itertools
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sys
from lib.envs.windy_gridworld import WindyGridworldEnv
matplotlib.style.use('ggplot')
```

We initialze an environment with 4 actions and one with 8 actions, where the latter includes the diagonal moves.

```python
env = WindyGridworldEnv(nA=4)
env2 = WindyGridworldEnv(nA=8)
```

In the code below you should complete the SARSA algorithm. The pseudocode for this algorithm is given on page 130 in the book by Sutton and Barto.

```python
def sarsa(env, num_episodes, discount_factor=1.0, alpha=0.5, epsilon=0.1):
    """
    SARSA algorithm: On-policy TD control. Finds the optimal epsilon-greedy policy.
    
    Args:
        env: Gymnasium environment.
        num_episodes: Number of episodes to run for.
        discount_factor: Gamma discount factor.
        alpha: TD learning rate.
        epsilon: Chance the sample a random action. Float betwen 0 and 1.
    
    Returns:
        A tuple (Q, stats).
        Q is the optimal action-value function, a dictionary mapping state -> action values.
        stats is an EpisodeStats object with two numpy arrays for episode_lengths and episode_rewards.
    """
    
    nA = env.action_space.n
    # The final action-value function.
    # A nested dictionary that maps state -> (action -> action-value).
    # Every key is a state, while the value is a numpy array of values per action of that state. 
    Q = {}
    for s1 in range(env.shape[0]):
        for s2 in range(env.shape[1]):
            Q[(s1, s2)] = np.zeros(nA)
        
    # Keeps track of useful statistics
    episode_rewards = np.zeros(num_episodes)
    episode_lengths = np.zeros(num_episodes)

    # Loop over the episodes
    for i in range(num_episodes):
        # Print every 100 episodes.
        if (i + 1) % 100 == 0:
            print("\rEpisode {}/{}.".format(i + 1, num_episodes), end="")
            sys.stdout.flush()
        
        # Reset the environment and pick the first action
        state = env.reset()
        
        # Calculate the action probabilities for the current state
        action_probs = np.ones(nA) * epsilon / nA
        best_action = np.argmax(Q[state])
        action_probs[best_action] += (1.0 - epsilon)
        
        # Sample an action from the action probabilities
        action = np.random.choice(np.arange(len(action_probs)), p=action_probs)
        
        # Perform one episode
        done = False
        t = 1
        while not done:
            # Take a step
            next_state, reward, done, _ = env.step(action)
                                    
            # Calculate the action probabilities for the next state
            next_action_probs = np.ones(nA) * epsilon / nA
            best_action = np.argmax(Q[next_state])
            next_action_probs[best_action] += (1.0 - epsilon)
            # Sample an action from the action probabilities
            next_action = np.random.choice(np.arange(len(action_probs)), p=next_action_probs)
            
            # Update statistics
            episode_rewards[i] += reward
            episode_lengths[i] = t
            t += 1

            # TD Update
            td_target = reward + discount_factor * Q[next_state][next_action]
            td_delta = td_target - Q[state][action]
            Q[state][action] += alpha * td_delta
                   
            # Update state and action
            action = next_action
            state = next_state        
            
    return Q, episode_rewards, episode_lengths
```

```python
# Determine if we run either the environment with 4 or 8 actions
env = env 
# Run SARSA
Q, episode_rewards, episode_lengths = sarsa(env, 1000)
```

```python
# Render the final policy
state = env.reset()
env.render()
done = False
while not done:
    action = np.argmax(Q[state])
    next_state, reward, done, _ = env.step(action)
    print("State:%s, action: %s, next state: %s" % (state, env.actions[action], next_state))
    env.render()
    state = next_state
```

```python
# Plot the episode length over time
fig1 = plt.figure(figsize=(10,5))
plt.plot(episode_lengths)
plt.xlabel("Episode")
plt.ylabel("Episode Length")
plt.title("Episode Length over Time")
plt.show(fig1)

# Plot the episode reward over time
fig2 = plt.figure(figsize=(10,5))
rewards_smoothed = pd.Series(episode_rewards).rolling(10, min_periods=10).mean()
plt.plot(rewards_smoothed)
plt.ylim(ymin=-100, ymax=0)
plt.xlabel("Episode")
plt.ylabel("Episode Reward (Smoothed)")
plt.title("Episode Reward over Time (Smoothed over window size {})".format(10))
plt.show(fig2)

# Plot time steps and episode number
fig3 = plt.figure(figsize=(10,5))
plt.plot(np.cumsum(episode_lengths), np.arange(len(episode_lengths)))
plt.xlabel("Time Steps")
plt.ylabel("Episode")
plt.title("Episode per time step")
plt.show(fig3)
```
