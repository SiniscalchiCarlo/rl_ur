# Maximization Bias and double Q-learning

In this exercise we examine the maximization bias of the Q-learning. We will replicate example 6.7 on page 134 and 135 in the book. The maximization step in Q-learning and the other control methods can cause overestimation of some action-values, which can cause the algorithm to often pick wrong actions. In this exercise we have given an implementation of Q-learning and you have to implement the double Q-learning algorithm, which circumvents the maximization bias issue.

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import matplotlib
import sys
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
```

Below we have implemented the environment that models example 6.7 from the book. We have chosen for state B to have 10 actions. You can play around with this number to examine the effects on the algorithms.

```python
class MaximizationBias(gym.Env):
    def __init__(self):
        # Initialize state and action space
        # We have 4 states, terminal state 0, state 1 (B), state 2 (A) and terminal state 3
        self.observation_space = spaces.Discrete(4)
        self.action_space = spaces.Discrete(10)
               
    def reset(self):
        # Start in s
        self.state = 2
        return self.state
    
    def step(self, action):
        # For state 2 (A in the example) we can only take action 0 (left) or 1 (right). Higher values will be set to 1.
        if self.state == 2:
            action = min(action, 1)
        
        # Determine the next state
        if self.state == 1:
            next_state = 0
            reward = np.random.normal(-0.1, 1)
            done = True
        elif self.state == 2 and action == 0:
            next_state = 1
            reward = 0
            done = False
        elif self.state == 2 and action == 1:
            next_state = 3
            reward = 0
            done = True
        else:
            raise ValueError('Cant take actions from terminal states')

        # Update state
        self.state = next_state
               
        return next_state, reward, done, {}
```

```python
def q_learning(env, num_episodes, discount_factor=1.0, alpha=0.1, epsilon=0.1):
    """
    Q-Learning algorithm: Off-policy TD control. Finds the optimal greedy policy
    while following an epsilon-greedy policy
    
    Args:
        env: Gymnasium environment.
        num_episodes: Number of episodes to run for.
        discount_factor: Gamma discount factor.
        alpha: TD learning rate.
        epsilon: Chance to sample a random action. Float between 0 and 1.
    
    Returns:
        A tuple (Q, episode_lengths).
        Q is the optimal action-value function, a dictionary mapping state -> action values.
        stats is an EpisodeStats object with two numpy arrays for episode_lengths and episode_rewards.
    """
    
    # The final action-value function.
    # A nested dictionary that maps state -> (action -> action-value).
    Q = {}
    Q[0] = np.zeros(1)
    Q[1] = np.zeros(env.action_space.n)
    Q[2] = np.zeros(2)
    Q[3] = np.zeros(1)

    # Keep track of the amount of left actions taken in A
    left_action = np.zeros(num_episodes)
    
    # Loop over episodes
    for i in range(num_episodes):       
        # Reset the environment and pick the first action
        state = env.reset()
        
        # One step in the environment
        done = False
        t = 1
        while not done:          
            # Calculate the action probabilities for the current state
            if state == 1:
                nA = env.action_space.n
            elif state == 2:
                nA = 2
                
            action_probs = np.ones(nA) * epsilon / nA
            best_actions = np.argwhere(Q[state] == np.amax(Q[state])).ravel()
            best_action = np.random.choice(best_actions)
            action_probs[best_action] += (1.0 - epsilon)

            # Sample an action from the action probabilities
            action = np.random.choice(np.arange(len(action_probs)), p=action_probs)
            
            # Increment left action
            if state == 2 and action == 0:
                left_action[i] = 1
    
            # Take a step
            next_state, reward, done, _ = env.step(action)

            # TD Update
            td_delta = reward + discount_factor * Q[next_state][np.argmax(Q[next_state])] - Q[state][action]
            Q[state][action] += alpha * td_delta
                
            # Update state
            state = next_state
    
    return Q, left_action
```

In the code cell above we have given a complete implementation of the Q-learning algorithm. Below you should implement double Q-learning, for which the pseudocode is given on page 136 in the book. To do this you have to adapt the Q-learning code that is given above.

```python
def double_q_learning(env, num_episodes, discount_factor=1.0, alpha=0.1, epsilon=0.1):
    """
    Q-Learning algorithm: Off-policy TD control. Finds the optimal greedy policy
    while following an epsilon-greedy policy
    
    Args:
        env: Gymnasium environment.
        num_episodes: Number of episodes to run for.
        discount_factor: Gamma discount factor.
        alpha: TD learning rate.
        epsilon: Chance to sample a random action. Float between 0 and 1.
    
    Returns:
        A tuple (Q, episode_lengths).
        Q is the optimal action-value function, a dictionary mapping state -> action values.
        stats is an EpisodeStats object with two numpy arrays for episode_lengths and episode_rewards.
    """
    
    # The final action-value function.
    # A nested dictionary that maps state -> (action -> action-value).
    Q1 = {}
    Q1[0] = np.zeros(1)
    Q1[1] = np.zeros(env.action_space.n)
    Q1[2] = np.zeros(2)
    Q1[3] = np.zeros(1)
    
    Q2 = {}
    Q2[0] = np.zeros(1)
    Q2[1] = np.zeros(env.action_space.n)
    Q2[2] = np.zeros(2)
    Q2[3] = np.zeros(1)

    # Keep track of the amount of left actions taken in A
    left_action = np.zeros(num_episodes)
    
    # Loop over episodes
    for i in range(num_episodes):       
        # Reset the environment and pick the first action
        state = env.reset()
        
        # One step in the environment
        done = False
        t = 1
        while not done:          
            # Calculate the action probabilities for the current state
            if state == 1:
                nA = env.action_space.n
            elif state == 2:
                nA = 2
                
            # Determine action probabilities
            action_probs = np.ones(nA) * epsilon / nA
            value_func = Q1[state] + Q2[state]
            best_actions = np.argwhere(value_func == np.amax(value_func)).ravel()
            best_action = np.random.choice(best_actions)
            action_probs[best_action] += (1.0 - epsilon)

            # Sample an action from the action probabilities
            action = np.random.choice(np.arange(len(action_probs)), p=action_probs)
            
            # Increment left action
            if state == 2 and action == 0:
                left_action[i] = 1
    
            # Take a step
            next_state, reward, done, _ = env.step(action)

            # TD Update
            if np.random.random() < 0.5:
                td_delta = reward + discount_factor * Q2[next_state][np.argmax(Q1[next_state])] - Q1[state][action]
                Q1[state][action] += alpha * td_delta
            else:
                td_delta = reward + discount_factor * Q1[next_state][np.argmax(Q2[next_state])] - Q2[state][action]
                Q2[state][action] += alpha * td_delta
            
            # Update state
            state = next_state
    
    return Q1, Q2, left_action
```

```python
env = MaximizationBias()
```

```python
num_runs = 10000
num_episodes = 300
percentage_left_normal = np.zeros(num_episodes)
percentage_left_double = np.zeros(num_episodes)
for r in range(num_runs):
    # Print every 100 runs.
    if (r + 1) % 100 == 0:
        print("\rEpisode {}/{}.".format(r + 1, num_runs), end="")
        sys.stdout.flush()
        
    # Run Q-learning and double Q-learning
    Q, left_action_normal = q_learning(env, num_episodes)
    Q1, Q2, left_action_double = double_q_learning(env, num_episodes)
    
    # Update left action percentage
    percentage_left_normal += left_action_normal
    percentage_left_double += left_action_double
    
percentage_left_normal = percentage_left_normal/num_runs
percentage_left_double = percentage_left_double/num_runs
```

```python
plt.plot(percentage_left_normal)
plt.plot(percentage_left_double)
plt.xlabel("Episodes")
plt.ylabel("%% left action from A")
plt.legend(["Standard Q-learning", "Double Q-learning"])
plt.title("Maximization bias example for {} runs.".format(num_runs))
plt.show()
```
